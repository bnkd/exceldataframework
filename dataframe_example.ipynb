```python
import pandas as pd
```

```python
base_df = pd.read_csv("consumer.csv")
detail_df = pd.read_csv("backend.csv")
base_df['items'] = base_df["items"].apply(lambda row : row.strip("[]''"))
base_df
```

```python
result_df = base_df.merge (detail_df[["items","cost_price", "HSN code", "Tax Slab ", "Supplier"]], 
                   how = "inner",left_on = "items", right_on = "items")
result_df.to_csv("rec.csv")
result_df
```
